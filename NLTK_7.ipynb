Read the text file `UN_100k.fr` and perform the following steps: first, convert all characters to lowercase to normalize the text. Next, tokenize the text and concatenate all the tokens into a single string with a space character added between each token to maintain separation. Finally, save this processed and tokenized text to a new file named `UN_100k_tokenized.fr`.

````python
from nltk import word_tokenize
TEXT_TOKENIZED = 'GloVe/UN_100k_tokenized.fr'

# Read the UN_100k file
with open('GloVe/UN_100k.fr', 'r', encoding = 'utf-8') as reader:
    lines = reader.readlines()

with open(TEXT_TOKENIZED, 'w', encoding = 'utf-8') as writer:
    for line in lines:
        # Normalize the text to lower case and tokenize accordingly
        tokens = word_tokenize(line.lower())

        # Concatenate the tokens into a single string
        new_line = ''
        for t in tokens:
            new_line = new_line + ' ' + t

        # Write the string into the file
        writer.write(new_line + '\n')
````
